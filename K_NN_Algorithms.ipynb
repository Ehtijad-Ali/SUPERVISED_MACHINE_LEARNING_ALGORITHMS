```python
import pandas as pd

# Importing the load_iris function from the sklearn.datasets module
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt

# Loading the Iris dataset using the load_iris() function and storing it in the variable iris
iris = load_iris()

```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
# Creating a DataFrame from the Iris dataset features, using iris.data as the data and iris.feature_names as the column names
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Adding a new column 'target'.
df['target'] = iris.target

df.head()
```

```python
df.shape
```

```python
# give 1 to the target value column.
df[df.target==1].head()
```

```python
# Adding a new column 'flower_name'
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])

df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### Sapel length vs Sapel width(Setosa vs Versicolor)

```python
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df[df['target'] == 0]['sepal length (cm)'], df[df['target'] == 0]['sepal width (cm)'], color='red', marker='+', label='Setosa')
plt.scatter(df[df['target'] == 1]['sepal length (cm)'], df[df['target'] == 1]['sepal width (cm)'], color='black', marker='.', label='Versicolor')
plt.legend()
plt.show()
```

```python
# Importing the train_test_split function from the sklearn.model_selection module
from sklearn.model_selection import train_test_split

# Extracting features (input variables) by dropping the 'target' and 'flower_name' columns from the DataFrame
x = df.drop(['target', 'flower_name'], axis='columns')

# Extracting the target variable (output variable) from the DataFrame
y = df.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
len(x_train)
```

```python
len(x_test)
```

### Create KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)
```

```python
# Importing the KNeighborsClassifier class from the sklearn.neighbors module
from sklearn.neighbors import KNeighborsClassifier

# Creating a KNeighborsClassifier object with n_neighbors set to 10
knn = KNeighborsClassifier(n_neighbors=10)

# Fitting the KNN model to the training data
knn.fit(x_train, y_train)
```

```python
# Calculating the accuracy score of the KNN model on the test data
knn.score(x_test, y_test)
```
